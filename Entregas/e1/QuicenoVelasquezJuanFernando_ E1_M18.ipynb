```
#'image' es una matriz de NumPy o una lista de listas
# con valores de píxeles en escala de grises (0-255)

#cargar la imagen
image = cv2.imread('/content/drive/MyDrive/2023-2/ML/machine_learning/Entregas/e1/entero.jpg')
plt.imshow(image)

# Inicializamos las variables
fil, col = 0, 0
b = 1

# Tamaño del cuadro en el tablero de ajedrez
tamano_cuadro = 50

# Recorremos la matriz 'image'
for i in range(len(image)):
    fil = (fil + 1) % tamano_cuadro  # Usamos el operador módulo para reiniciar el contador
    if fil == 0:
        b = 3 - b  # Cambiamos el valor de la bandera (1 se convierte en 2, y 2 en 1)

    for j in range(len(image[i])):
        col = (col + 1) % tamano_cuadro  # Usamos el operador módulo para reiniciar el contador
        if col == 0:
            b = 3 - b  # Cambiamos el valor de la bandera

        if b == 1:
            image[i][j] += 255  # La imagen es de 8 bits (0-255)
# Muestre la imagen y guardela en drive usando opencv (consultar)
print(image.shape)
plt.imshow(image)
#guardar imagen
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/machine learning/Entregas/e1/tablero.png',image)
```

```
# Lea la imagen
image = cv2.imread('/content/drive/MyDrive/2023-2/ML/machine_learning/Images/lena.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #pasar a formato rgb
plt.imshow(image)
```

```
#primero obtiene la zona de la imagen que vamos a replicar
image_cut=image[90:190,170:270,0:3] #guardar imagen recortada
print(image_cut.shape)
plt.imshow(image_cut)
```

```
#generar la imagen resultado
img=image.copy()
k=0
l=0

#ciclo para recorrer la imagen y copiar la imagen cortada en ella
for i in range(0,300):
  for j in range(0,400):
    if l<99: #contador para recorrer columnas de la imagen cortada
      l+=1
    else:
      l=0
    img[i,j]= image_cut[k,l]
  if k<99: #contador para recorrer filas de la imagen cortada
    k+=1
  else:
    k=0

# Muestre la imagen y guardela en drive usando opencv (consultar)
print(img.shape)
plt.imshow(img)
#para evitar el cambio de color se debe pasar formato
cv2.imwrite('/content/drive/MyDrive/2023-2/ML/machine_learning/Entregas/e1/varias.jpg', cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```
